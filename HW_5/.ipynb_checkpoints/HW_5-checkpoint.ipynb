## Задание 1.
Реализация базового класса

```python
from datetime import datetime
from typing import List, Any
import json
from decimal import Decimal
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```python
# прим. в задании указано использовать тип float, но я предпочла Decimal, чтобы избежать нюансов с поведением 
# float-а в мат. операциях 
# это касается случаев, когда 200.0 - 150.55 дает = 49.44999999999999

class CustomError(Exception):
    """Кастомное исключение для операций со счетом"""
    pass


class Account:
    """Класс банковского счета

    Атрибуты:
        holder: Имя владельца счета
        account_number: Уникальный номер счета
        _balance: Текущий баланс счета
        operations_history: История операций по счету
    """

    _account_counter: int = 1000

    def __init__(self, account_holder: str, balance: Decimal = Decimal('0.0')) -> None:
        """Инициализация счета с начальным нулевым баланслм

        Аргументы:
            account_holder: Имя владельца счета
            balance: Начальный баланс
        """
        self._validate_holder_name(account_holder)
        
        self.holder: str = account_holder
        self.account_number: str = f'ACC-{Account._account_counter}'
        self._balance: Decimal = balance
        self.operations_history: List[Dict[str, Any]] = []

        Account._account_counter += 1

        self.add_operation(
            'Account opened with started balance',
            self._balance,
            'success'
        )
        
    @staticmethod
    def _validate_holder_name(holder_name: str) -> None:
        """Проверка корректности имени владельца счета
           на кириллице и латинице

        Аргументы:
            name: Имя владельца счета

        Исключения:
            CustomError: Если имя владельца не соответствует формату
        """
        cyrylic = r'^[А-ЯЁ][а-яё]+ [А-ЯЁ][а-яё]+$'
        latin = r'^[A-Z][a-z]+ [A-Z][a-z]+$'
        if not (re.match(cyrylic, holder_name) 
                or re.match(latin, holder_name)):
            raise CustomError(
                'Holder name must be "Имя Фамилия" or "Name Surname" with camel case'
            )

    def add_operation(self, description: str, amount: Decimal,
                      status: str, reason: str = None) -> None:
        """Сохранение операции в истории операций

        Аргументы:
            description: Тип операции (deposit / withdraw)
            amount: Сумма операции
            status: Статус операции (success / failed)
            reason: В случае статуса failed, причина отклонения
        """
        transaction = {
            'description': description,
            'amount': amount,
            'balance after operation': self._balance,
            'operation date': datetime.now().strftime('%d/%m/%Y %H:%M:%S'),
            'status': status,
            'reason': reason
        }
        self.operations_history.append(transaction)

    def deposit(self, amount: Decimal) -> None:
        """Операция пополнения счета

        Аргументы:
            amount: Сумма пополнения счета 

        Исключения:
            CustomError: Если сумма введена некорректно,
                         меньше или равна нулю
        """
        if isinstance(amount, Decimal):
            if amount > 0:
                self._balance += amount
                self.add_operation('deposit', amount, 'success')
            else:
                error_type = 'Deposit amount can not be negative'
                self.add_operation('deposit', amount, 'failed', error_type)
                raise CustomError(error_type)
        else:
            error_type = 'Incorrect deposit amount'
            raise CustomError(error_type)

    def withdraw(self, amount: Decimal) -> None:
        """Операция сиписания со счета

        Аргументы:
            amount: Сумма списания со счета

        Исключения:
            CustomError: Если сумма введена некорректно,
                         меньше или равна нулю
                         или больше остатка на счете
        """
        if isinstance(amount, Decimal):
            if amount > 0:
                if amount < self._balance:
                    self._balance -= amount
                    self.add_operation('withdraw', -amount, 'success')
                else:
                    error_type = 'Not enough money on account'
                    self.add_operation('withdraw', -amount, 'failed', error_type)
                    raise CustomError(error_type)
            else:
                error_type = 'Withdrawal amount can not be negative'
                self.add_operation('withdraw', -amount, 'failed', error_type)
                raise CustomError(error_type)
        else:
            error_type = 'Incorrect withdrawal amount'
            raise CustomError(error_type)

    def get_operation_history(self) -> str:
        """Возвращает список операций по счету в формате JSON
        """
        def _as_str(o: Any) -> str:
            if isinstance(o, Decimal):
                return str(o)
            raise TypeError

        return json.dumps(
            self.operations_history,
            ensure_ascii=False,
            indent=2,
            default=_as_str,
        )

    def get_balance(self) -> Decimal:
        """Возвращает текущий баланс счета
        """
        return self._balance

    def plot_history(self) -> pd.DataFrame:
        """Построение графика динамики движения по счету

        Возвращает:
            pd.DataFrame: Датафрейм с историей операций

        Исключения:
            CustomError: Если нет операций по счету
        """
        if not self.operations_history:
            raise CustomError('No operations for showing')

        df = pd.DataFrame(self.operations_history)
        df['operation date'] = pd.to_datetime(
            df['operation date'],
            format='%d/%m/%Y %H:%M:%S',
            dayfirst=True,
            errors='coerce',
        )
        df = df.sort_values('operation date')

        def _to_float(x: Any) -> Any:
            return float(x) if isinstance(x, Decimal) else x

        df['balance after operation'] = df['balance after operation'].apply(_to_float)
        df['amount'] = df['amount'].apply(_to_float)

        fig, ax = plt.subplots(figsize=(9, 5))
        ax.plot(
            df['operation date'],
            df['balance after operation'],
            marker='o',
            linewidth=2,
            color='#963ff3',
        )

        title = (
            f'Balance changing dynamics for {self.account_number} '
            f'({self.holder})'
        )
        ax.set_title(title)
        ax.set_xlabel('Operation time')
        ax.set_ylabel('Balance after operation')
        ax.grid(True, linestyle='--', alpha=0.4)
        ax.tick_params(axis='x', rotation=30)
        fig.tight_layout()
        plt.show()

        return df


    def analyze_operations(self, top: int = 3) -> str:
        """Возвращает последние крупные операции по дате
        По умолчанию - 3

        Аргументы:
            top: Количество возвращаемых операций

        Возвращает:
            str: JSON-строка с последними крупными операциями

        Исключения:
            CustomError: Если нет операций по счету
        """
        def _as_str(o: Any) -> str:
            if isinstance(o, Decimal):
                return str(o)
            raise TypeError
            
        def parse_dt(dt_s: str) -> datetime:
            return datetime.strptime(dt_s, '%d/%m/%Y %H:%M:%S')
            
        if not self.operations_history:
            raise CustomError('No operations to analyze')
            
        sorted_operations = sorted(
            self.operations_history,
            key=lambda op: (abs(op['amount']), parse_dt(op['operation date'])),
            reverse=True,
        )
        top_operations = sorted_operations[:max(0, int(top))]

        return json.dumps(top_operations, ensure_ascii=False, indent=2, default=_as_str)
```

### Задание 2

```python
class CheckingAccount(Account):
    """Класс обычного расчетного счета
    """
    account_type: str = 'checking'


class SavingsAccount(Account):
    """Класс сберегательного счета"""
    account_type: str = 'savings'

    def apply_interest(self, rate: Decimal) -> None:
        """Начисление процентов на остаток по счету

        Аргументы:
            rate: Размер ставки (либо в долях, либо в процентах)

        Исключения:
            CustomError: Если ставка введена некорректно 
                         или меньше или равна нулю
        """
        if isinstance(rate, Decimal):
            if rate > 0:      
                effective_rate = rate / Decimal('100') if rate > 1 else rate
                interest = self._balance * effective_rate
                self._balance += interest
                self.add_operation('apply interest', interest, 'success')
            else:
                raise CustomError('Rate amount can not be negative')
        else:
            raise CustomError('Incorrect rate amount')
      
    def withdraw(self, amount: Decimal) -> None:
        """Списание средств со счета с учетом ограничения на 50% от суммы

        Аргументы:
            amount: Сумма списания со счета

        Исключения:
            CustomError: Если сумма введена некорректно,
                         меньше или равна нулю
                         или больше 50% от суммы остатка
        """
        if not isinstance(amount, Decimal):
            raise CustomError('Incorrect withdrawal amount')

        if amount <= 0:
            error_type = 'Withdrawal amount can not be negative'
            self.add_operation('withdraw', -amount, 'failed', error_type)
            raise CustomError(error_type)

        half_balance = (self._balance * Decimal('0.5'))
        if amount > half_balance:
            error_type = 'Withdrawal amount is more than 50% of balamce'
            self.add_operation('withdraw', -amount, 'failed', error_type)
            raise CustomError(error_type)

        super().withdraw(amount)
```
