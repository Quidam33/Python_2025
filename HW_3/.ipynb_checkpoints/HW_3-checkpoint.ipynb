```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Electronic_sales_Sep2023-Sep2024.csv')
df.head()
```

# Задание 1

### Общие траты по каждому покупателю

```python
df['Order Status'].unique()
```

```python
df[df['Order Status'] != 'Cancelled'].groupby('Customer ID').apply(lambda x: (x['Total Price'] + x['Add-on Total']).sum(), include_groups=False).reset_index(name='Total_Sum')
```

### Количество денег на аксессуары и доп услуги

```python
df[df['Order Status'] != 'Cancelled'].groupby('Customer ID')['Add-on Total'].sum()
```

### Предпочитаемый метод оплаты

```python
df[df['Order Status'] != 'Cancelled'].groupby(['Customer ID', 'Payment Method']).count()
```

# Задание 2

```python
df_shipping_type = df[df['Order Status'] != 'Cancelled'].groupby('Shipping Type').apply(lambda x: (x['Total Price'] + x['Add-on Total']).sum(), include_groups=False).reset_index(name='Total_Sum')
df_shipping_type
```

```python
colors = ['red', 'blue', 'green', 'purple', 'orange']
df_shipping_type.plot(x='Shipping Type', y='Total_Sum', kind='bar', color=colors, title='По типу доставки')
plt.ylabel('Total_Sum')
plt.show()
```

```python
df_product_type = df[df['Order Status'] != 'Cancelled'].groupby('Product Type')['Total Price'].sum()
df_product_type
```

```python
colors = ['red', 'blue', 'green', 'purple', 'orange']
df_product_type.plot(x='Product Type', y='Total Price', kind='bar', color=colors, title='По типу продукта')
plt.ylabel('Total Price')
plt.show()
```

```python
df['Purchase Date'] = pd.to_datetime(df['Purchase Date'])

df['Purchase Month'] = df['Purchase Date'].dt.strftime('%Y-%m-01')
df.head()
```

```python
df_by_mn = df[df['Order Status'] != 'Cancelled'].groupby('Purchase Month')['Add-on Total'].sum()
df_by_mn
```

```python
df_by_mn.sort_index(inplace=True) 
plt.figure(figsize=(15, 6))
plt.plot(df_by_mn.index, df_by_mn.values, marker='o')
plt.show()
```

```python
df['Purchase Quarter'] = df['Purchase Date'].dt.to_period('Q').dt.start_time
df.head()
```

```python
df_by_qrt = df[df['Order Status'] != 'Cancelled'].groupby('Purchase Quarter')['Add-on Total'].sum()
df_by_qrt.head()
```

```python
df_by_qrt.sort_index(inplace=True) 
plt.figure(figsize=(8, 6))
plt.plot(df_by_qrt.index, df_by_qrt.values, marker='o')
plt.show()
```
